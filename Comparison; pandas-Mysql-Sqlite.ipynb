### Pandas & csv

```python
def csv_query(variable, number):
    complaints = []
    chunksize = 10 ** 6
    fields = ['Agency Name','Complaint Type','Borough']
    i=0
    for chunk in pd.read_csv('service_requests.csv', chunksize=chunksize,usecols=fields):
        if variable == 'complaint': 
            complaints.extend(chunk['Complaint Type'])
        if i==0:
            if variable == 'borough':
                by_borough = (chunk.groupby('Borough')['Complaint Type']
                         .value_counts()
                         .rename_axis(['borough','complaint'])
                         .reset_index(name='freq'))
            if variable == 'agency':
                by_agency = (chunk.groupby('Agency Name')['Complaint Type']
                         .value_counts()
                         .rename_axis(['Agency','complaint'])
                         .reset_index(name='freq'))
        else:
            if variable == 'borough':
                by_borough = by_borough.append((chunk.groupby('Borough')['Complaint Type']
                             .value_counts()
                             .rename_axis(['borough','complaint'])
                             .reset_index(name='freq')))
            if variable == 'agency':
                by_agency = by_agency.append((chunk.groupby('Agency Name')['Complaint Type']
                         .value_counts()
                         .rename_axis(['Agency','complaint'])
                         .reset_index(name='freq')))
        if variable == 'borough':
            by_borough = by_borough.groupby(['borough','complaint'])['freq'].sum().reset_index()
        if variable == 'agency':    
            by_agency = by_agency.groupby(['Agency','complaint'])['freq'].sum().reset_index()
        i+=1
    if variable == 'complaint': 
        most_common = Counter(complaints)
        return most_common.most_common()
    if variable == 'borough':
        borough_complaint = pd.DataFrame(columns = by_borough.columns)
        for i in by_borough['borough'].unique():
            borough_complaint = borough_complaint.append(by_borough.loc[by_borough['borough'] == i].sort_values('freq',ascending=False))
        bb=[]
        for i in borough_complaint['borough'].unique():
            bb.append(borough_complaint.loc[borough_complaint['borough'] == i].head(number))
        return bb
    if variable == 'agency':   
        agency_complaint = pd.DataFrame(columns = by_agency.columns)
        for i in by_agency['Agency'].unique():
            agency_complaint = agency_complaint.append(by_agency.loc[by_agency['Agency'] == i].sort_values('freq',ascending=False))
        bb= []
        bb = agency_complaint.sort_values('freq',ascending=False).head(number)
        #for i in agency_complaint['Agency'].unique():
         #   bb.append(agency_complaint.loc[agency_complaint['Agency'] == i].head(number))
        return bb
```

```python
%%time
csv_query('complaint',1)
```

```python
%%time
csv_query('borough',5)
```

```python
%%time
csv_query('agency',25)
```

### MySQL 

```python
import pyodbc
conn = pyodbc.connect('Driver={SQL Server Native Client 11.0};'
                      'Server=localhost;'
                      'Database=python;'
                      'Trusted_Connection=yes;')
cursor = conn.cursor()
cursor.execute('''
 CREATE TABLE Service5 (
        Agency_Name VARCHAR(100) ,
        Complaint_Type VARCHAR(100) ,
        Borough  VARCHAR(100) 
 
);
               ''')
conn.commit()
```

```python
%%time
fields = ['Agency Name','Complaint Type','Borough']
chunksize = 10 ** 6
for chunk in pd.read_csv('service_requests.csv', chunksize=chunksize,usecols=fields):
    for row in chunk.itertuples():
        if type(row._1) == float:
            row1 = ""
        else:
            if len(row._1) <=4:
                row1 = ""
            else:
                row1 = row._1
        if type(row._2) == float:
            row2 = ""
        else:
            if len(row._2) <=4:
                row2 = ""
            else:
                row2 = row._2
        if type(row.Borough) == float:
            rowBorough = ""
        else:
            if len(row.Borough) <=2:
                rowBorough = ""
            else:
                rowBorough = row.Borough
        cursor.execute('''
                    INSERT INTO Service5 (Agency_Name, Complaint_Type, Borough)
                    VALUES (?,?,?)
                    ''',
                    row1, 
                    row2,
                    rowBorough
                    )
    conn.commit()
```

```python
%%time
cursor.execute('''
 SELECT Complaint_Type, COUNT(Complaint_Type)
     FROM Service5
     GROUP BY Complaint_Type
     ORDER BY COUNT(*) DESC
               ''')
result=cursor.fetchall()
result
```

```python
%%time
cursor.execute('''
 SELECT Borough, Complaint_Type, COUNT(Complaint_Type) as ct
     FROM Service5
     WHERE Borough != 'Unspecified'
     GROUP BY Borough, Complaint_Type
     Having COUNT(Complaint_Type) > 65000
     ORDER BY Borough DESC, COUNT(*) DESC
               ''')
result_b=cursor.fetchall()
result_b
```

```python
%%time
cursor.execute('''
 SELECT Agency_Name, Complaint_Type, COUNT(Complaint_Type)
     FROM Service5
     WHERE Agency_Name != ''
     GROUP BY Agency_Name, Complaint_Type
     Having COUNT(Complaint_Type) > 10000
     ORDER BY COUNT(Complaint_Type) DESC
               ''')
result_a=cursor.fetchall()
result_a
```

### Efficiency improvement
- Seeking through only essenial columns (done before)
- While converting to SQL database drop rows with blanks/redundant information
- limiting the output 


```python
import sqlite3
connect = sqlite3.connect('python.db')
curs = connect.cursor()
curs.execute('''
 CREATE TABLE Service7 (
        Agency_Name VARCHAR(100) ,
        Complaint_Type VARCHAR(100) ,
        Borough  VARCHAR(100) 
 
);
               ''')
connect.commit()
```

```python
%%time
fields = ['Agency Name','Complaint Type','Borough']
chunksize = 10 ** 6
for chunk in pd.read_csv('service_requests.csv', chunksize=chunksize,usecols=fields):
    for row in chunk.itertuples():
        if row._1 =="" or row._2=="" or row.Borough=="":
            continue
        if type(row._1) == float:
            continue
        else:
            if len(row._1) <=1:
                continue
            else:
                row1 = row._1
        if type(row._2) == float:
            continue
        else:
            if len(row._2) <=1:
                continue
            else:
                row2 = row._2
        if type(row.Borough) == float or row.Borough == 'Unspecified':
            continue
        else:
            if len(row.Borough) <=2:
                continue
            else:
                rowBorough = row.Borough
        curs.execute('''
                    INSERT INTO Service7 (Agency_Name, Complaint_Type, Borough)
                    VALUES (?,?,?)
                    ''',
                    (row1, 
                    row2,
                    rowBorough)
                    )
    connect.commit()
```

```python
%%time
curs.execute('''
 SELECT Complaint_Type, COUNT(Complaint_Type)
     FROM Service7
     GROUP BY Complaint_Type
     Having COUNT(Complaint_Type) >100000
     ORDER BY COUNT(*) DESC
               ''')
result=curs.fetchall()
result
```
