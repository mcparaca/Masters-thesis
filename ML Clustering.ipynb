```python
import pandas as pd
import openpyxl
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import AgglomerativeClustering
```

```python
data = pd.read_csv('wine.csv', engine='c')
```

## K means

```python
df_dropped = data.drop('quality', axis=1)
X =pd.DataFrame(df_dropped.values[:, 1:])
Clus_dataset = StandardScaler().fit_transform(X)

from sklearn.preprocessing import MaxAbsScaler
transformer = MaxAbsScaler().fit(X)
X = transformer.transform(X)
X =pd.DataFrame(X)
```

```python
def plot2D(k, X):
    %matplotlib inline
    kmeans = KMeans(n_clusters=k, random_state=0).fit(X)
    k_means_labels = kmeans.labels_  
    f3, ax = plt.subplots(figsize=(13, 9))
    plt.scatter(X[1], X[4], c=k_means_labels.astype(np.float))
    plt.xlabel('alcohol', fontsize=18)
    plt.ylabel('total sulfur dioxide', fontsize=16)
    plt.title('Clusters in 2D plot')
    plt.show()
```

```python
plot2D(2, X)
```

```python
plot2D(3, X)
```

```python
plot2D(6, X)
```

```python
def plot3D(k, X):
    %matplotlib qt
    kmeans = KMeans(n_clusters=k, random_state=0).fit(X)
    k_means_labels = kmeans.labels_ 
    f4 = plt.figure(1, figsize=(13, 9))
    plt.clf()
    ax = Axes3D(f4, rect=[0, 0, .95, 1], elev=48, azim=134)
    plt.cla()
    ax.set_xlabel('alcohol')
    ax.set_ylabel('total sulfur dioxide')
    ax.set_zlabel('pH')
    ax.set_title('Clusters in 3D plot')
    ax.scatter(X[1], X[4], X[9], c= k_means_labels.astype(np.float))
```

```python
plot3D(3, X)
```

```python
plot3D(5, X)
```

```python
plot3D(7, X)
```

## Hierarchical Clustering

```python
def plot2D_hierarchical(k, X, link):
    %matplotlib inline
    hierarchical = AgglomerativeClustering(n_clusters=k, affinity='euclidean', linkage=link).fit(X)
    hierarchical_labels = hierarchical.labels_ 
    f3, ax = plt.subplots(figsize=(13, 9))
    plt.scatter(X[9], X[5], c=hierarchical_labels.astype(np.float))
    plt.xlabel('alcohol', fontsize=18)
    plt.ylabel('total sulfur dioxide', fontsize=16)
    plt.title('Clusters in 2D plot')
    plt.show()
```

```python
plot2D_hierarchical(4, X, 'ward')
```

```python
plot2D_hierarchical(5, X, 'average')
```

```python
plot2D_hierarchical(8, X, 'single')
```

```python
plot2D_hierarchical(5, X, 'complete')
```

```python
def plot3D_hierarchical(k, X, link):
    %matplotlib qt
    hierarchical = AgglomerativeClustering(n_clusters=k, affinity='euclidean', linkage=link).fit(X)
    hierarchical_labels = hierarchical.labels_ 
    f4 = plt.figure(1, figsize=(13, 9))
    plt.clf()
    ax = Axes3D(f4, rect=[0, 0, .95, 1], elev=48, azim=134)
    plt.cla()
    ax.set_xlabel('alcohol')
    ax.set_ylabel('total sulfur dioxide')
    ax.set_zlabel('pH')
    ax.set_title('Clusters in 2D plot')
    ax.scatter(X[4], X[5], X[7], c= hierarchical_labels.astype(np.float))
```

```python
plot3D_hierarchical(4, X, 'ward')
```

```python
plot3D_hierarchical(3, X, 'average')
```

```python
plot3D_hierarchical(6, X, 'single')
```
